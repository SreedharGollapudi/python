```python
## Transformation 

import numpy as np 

df['Log_Price'] = df['price'].apply(lambda x : np.log(x))

df.head()
```

```python
df['Normalized_Price'] = (df['price'] - df['price'].min()) / (df['price'].max()-df['price'].min())
df.head()
```

```python
# Data binning 

bin_edge = [0,100,200,300,400,500,np.inf]
bin_labels = ['0-100','101-200','201-300','301-400','401-500', '501+']
df['Price_Category'] = pd.cut(df['price'], bins = bin_edge, labels = bin_labels, right = False)
df.head()
```

```python
## Handling outliers 

from scipy.stats.mstats import winsorize

df['Winsoried_price'] = winsorize(df['price'], limits = [0.05,0.05])
df.head()
```
